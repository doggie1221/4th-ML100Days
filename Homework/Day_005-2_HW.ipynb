```python
import pandas as pd
import numpy as np
import urllib.request
import cv2

raw_data = pd.read_table("https://raw.githubusercontent.com/vashineyu/slides_and_others/master/tutorial/examples/imagenet_urls_examples.txt", header= None)
raw_data.columns=["ID","URL"]
print(raw_data)
```

```python
for i in range(5):
    req = urllib.request.urlopen(raw_data['URL'][i])
    image = np.asarray(bytearray(req.read()), dtype="uint8")
    image = cv2.imdecode(image, cv2.IMREAD_COLOR)
    print(image)
```
